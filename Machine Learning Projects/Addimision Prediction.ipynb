```python
# Logistic regression(Binary classification)Example
import numpy as np
import pandas as pd            
import matplotlib.pyplot as plt                                                                              
```

We are importing a dataset which include marks of two subject of applicants for admission in a university which are labelled as 1 for getting admission   and as 0 for not getting admission.

```python
data=pd.read_csv('ex2data1.txt',header=None)
data.head(6)
```

```python
#Seperating Data
data1=data[data[2]==1]
data2=data[data[2]==0]
```

```python
#data2
data1
```

DATA VISUALIZATION

##### Seperating data in two group: one for all Y=0 and another for all Y=1 and Plotting 

```python
data1=data[data[2]==1]
data2=data[data[2]==0]
plt.scatter(data1[0],data1[1],color='K',s=25,marker='*') #All the data with Y=1,represented by Black,Admitted
plt.scatter(data2[0],data2[1],color='R',s=25,marker='o') # All data with Y=0 ,represented by Red,Rejected
plt.title('Visualization of Data',fontsize=12,style='italic',fontweight='bold')
plt.xlabel('Marks in first subject',fontsize=12)
plt.ylabel('Marks in second subject',fontsize=12)
plt.show()
```

```python
data1
```

### Anohter way,  it can be plotted.

```python
X=data.iloc[:,[0,1]]
X
```

```python
Y=data.iloc[:,2]
Y
```

```python
plt.scatter(X[0],X[1],c=Y)
plt.title('Visualization of Data',fontsize=12,style='italic',fontweight='bold')
plt.xlabel('Marks in first subject',fontsize=12)
plt.ylabel('Marks in second subject',fontsize=12)
plt.show()
```

### Logistic Regression Library
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

MODEL BUILDING

```python
from sklearn.linear_model import LogisticRegression as lr
```

```python
clf = lr(random_state=42, solver='lbfgs',multi_class='multinomial')  #solver=lbfgh is gradient decent
clf.fit(X,Y)
```

```python
X
```

```python
Y
```

```python
y_pred=clf.predict(X)
y_pred
```

```python
clf.score(X, Y)  #checking accuracy
```

```python
from sklearn.metrics import mean_squared_error  #mean square error
mean_squared_error(Y,y_pred) 
```

```python
from math import sqrt    #Root mean square error
rms = sqrt(mean_squared_error(Y, y_pred))
rms
```

```python
clf.coef_ 
```

```python
clf.coef_[0,0]  #slicing 0,0 column
```

```python
clf.intercept_
```

```python
clf.intercept_[0]  #upper is in array slicing only numbers 
```

```python
#PLOTTING Decision Boundary
Y1=(-(clf.intercept_[0])-clf.coef_ [0,0]*X[0])/clf.coef_ [0,1]  #answer of y=mx+c is in ax+by+c=0 form so extracting y=(-c-ax)/b 
plt.plot(X[0],Y1)
########Plotting given data
plt.scatter(X[0],X[1],c=Y)
plt.title('Visualization of Data',fontsize=12,style='italic',fontweight='bold')
plt.xlabel('Marks in first subject',fontsize=12)
plt.ylabel('Marks in second subject',fontsize=12)
plt.show()
```
