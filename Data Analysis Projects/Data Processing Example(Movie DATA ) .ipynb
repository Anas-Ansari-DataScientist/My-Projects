```python
import pandas as pd
```

```python
pd.options.display.max_rows=10
unames=['user_id','gender','age','occupation','zip']
users=pd.read_table(r'D:\jupyter notebook\Dataset for Processing\Movie Data\users.dat',sep='::',header=None,names=unames)
rnames=['user_id','movie_id','rating','timestamp']
ratings=pd.read_table(r'D:\jupyter notebook\Dataset for Processing\Movie Data\ratings.dat',sep='::',header=None,names=rnames)
mnames=['movie_id','title','genres']
movies=pd.read_table(r'D:\jupyter notebook\Dataset for Processing\Movie Data\movies.dat',sep='::',header=None,names=mnames)
```

```python
u=users[:5]
u
```

```python
data=pd.merge(pd.merge(ratings,users),movies)
data
```

```python
mean_rating=data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')   #pivot_table always takes the mean automatically if we do not put aggfunc=mean it automatically gives mean 
mean_rating[:5]
```

```python
ratings_by_title=data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles=ratings_by_title.index[ratings_by_title>=250]
active_titles
```

```python
mean_ratings=mean_rating.loc[active_titles]
mean_ratings
```

```python
top_female_ratings=mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings
```

```python
top_male_ratings=mean_ratings.sort_values(by='M',ascending=False)
top_male_ratings
```

Measuring Rating Disagreement

```python
mean_ratings['diff']=mean_ratings['M']-mean_ratings['F']
mean_ratings
```

```python
sorted_by_diff=mean_ratings.sort_values(by='diff') #ascending=False
sorted_by_diff
```

```python
rating_std_by_title=data.groupby('title')['rating'].std()
rating_std_by_title
```

```python
rating_std_by_title=rating_std_by_title.loc[active_titles]
```
