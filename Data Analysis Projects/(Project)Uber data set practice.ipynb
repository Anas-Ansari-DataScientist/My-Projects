```python
#Importing Some Usefull Libraries 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
```

```python
#Impoerting Data

data=pd.read_csv('uber-rides-dataset.csv')
data.head(2)
```

```python
corrmat = data.corr() 
  
f, ax = plt.subplots(figsize =(9, 8)) 
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1) 
```

# RIDES COMPLETED/CANCELLED RATIO

```python
#Number of trip completed
a1=data['trip_status']=='Completed'
data.loc[a1]
c=data[a1].iloc[:,1]
com=c.count()
com
```

```python
#Number of trip Cancelled
a1=data['trip_status']=='Cancelled'
data.loc[a1]
c=data[a1].iloc[:,1]
can=c.count()
can
```

```python
###Total number of trip
a1=data['trip_status']=='Completed'
tot=a1.count()
tot
```

```python
#Percentage of trip completed
Percentage=(com/tot)*100
Percentage
```

```python
#Percentageof trip cancelled
Percentage=(can/tot)*100
Percentage
```
